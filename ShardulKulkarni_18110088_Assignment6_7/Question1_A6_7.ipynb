```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A=[0.4,0.06,0.1]
B=[0.4,0.01,0.1]

t0=0;
tf=1;
```

```python
M=np.matrix([[1,t0,t0**2,t0**3],[0,1,2*t0,3*(t0**3)],[1,tf,tf**2,tf**3],[0,1,2*tf,3*(tf**3)]])
M
```

```python
Y=np.matrix([[A[1],0,B[1],0]]).T
Y
```

```python
a=np.matmul(np.linalg.inv(M),Y)
a
```

```python
y=[]
vy=[]
ay=[]
t=[]
while (t0<=tf) :
  y.append(a[0,0]+a[1,0]*t0+a[2,0]*(t0**2)+a[3,0]*(t0**3))
  vy.append(a[1,0]+2*a[2,0]*t0+3*a[3,0]*(t0**2))
  ay.append(2*a[2,0]+6*a[3,0]*(t0))
  t.append(t0)
  t0+=0.01
```

```python
plt.plot(t,y)
plt.title("Position Vs Time")
plt.xlabel("Time")
plt.ylabel("Position y")
plt.grid()
plt.show()

plt.plot(t,vy)
plt.title("Velocity Vs Time")
plt.xlabel("Time")
plt.ylabel("Velocity y")
plt.grid()
plt.show()

plt.plot(t,ay)
plt.title("Acceleration Vs Time")
plt.xlabel("Time")
plt.ylabel("Acceleration y")
plt.grid()
plt.show()
```
